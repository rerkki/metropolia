```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dat = pd.read_csv("https://raw.githubusercontent.com/rerkki/metropolia/master/hytekapea.csv")

dat[0:10] #tulostetaan otsikko ja ensimm√§inen rivi tarkistuksen vuoksi
```

```
dat2 = dat[dat.Kysymys=='Koen usein rauhan tunnetta']
dat2
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(16,8)})

ax = sns.barplot(x=dat.Nimimerkki, y=dat2.Arvo, data=dat2)
```

```
dat3 = dat[dat.Nimimerkki==123098]
dat3
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(24,8)})

ax = sns.barplot(x=dat.Kysymys, y=dat3.Arvo, data=dat3)
```

```
fiilis = dat[dat.Kysymys=='Koen usein rauhan tunnetta']
fiilis_ = dat3[dat3.Kysymys=='Koen usein rauhan tunnetta']
teema = dat[dat.Kysymys=='Tunnen hallitsevani omaa elamaani'] 

Fiilis = pd.Series(fiilis.Arvo)
Teema = pd.Series(teema.Arvo)

from numpy.polynomial.polynomial import polyfit
b, m = polyfit(Fiilis, Teema, 1)

plt.plot(Fiilis, Teema, '.')
plt.plot(Fiilis, b + m * Fiilis, '-')
plt.title("Teema vs. Fiilis")
plt.show()
```
